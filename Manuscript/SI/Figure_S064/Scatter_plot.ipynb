```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('pyGAPS_results_SA.csv')
print(df.head(20)) 
```

```python
fig = plt.figure()
ax = plt.subplot(111)

count = 0
ax.plot([0, 600], [0, 600], c= 'gray')

for color in ['r','g', 'b']:
    for symbol in ['o', '*', 'D', 'P']:
        print(df.isotherm[count])
        ax.scatter(df.isotherm[count], df.kinetics[count],
                   c=color, 
                   marker=symbol, 
                   label = '{0}'.format(df.Sorbent[count] ))
        count = count + 1
        if count == 8: break
    if count == 8: break
ax.set_xlabel('BET surface area for equilibrium-based isotherm, m2/g')
ax.set_ylabel('BET surface area for SKID-based isotherm, m2/g')
ax.legend(bbox_to_anchor=(1.1, 1.05))
plt.show()
```
