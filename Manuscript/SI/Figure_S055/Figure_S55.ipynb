```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../../main_text/Figure_3/result_analysis_uptake.csv')
print(df.head()) 
```

```python
fig = plt.figure()
ax = plt.subplot(111)

count = 0
ax.plot([0, 50], [0, 50], c= 'gray')

for color in ['r','orange', 'magenta', 'g', 'b', 'cyan',  ]:
    for symbol in ['o', 's', '*', 'D', 'P']:
        print(df.isotherm_uptake[count])
        ax.scatter(df.isotherm_uptake[count], df.kinetics_uptake[count],
                   c=color, 
                   marker=symbol, 
                   label = '{0}, {1}% RH'.format(df.name[count], str(int(df.isotherm_RH[count])) ))
        count = count + 1
ax.set_xlabel('Uptake at RHmax for equilibrium-based isotherm, wt.%')
ax.set_ylabel('Uptake at RHmax for SKID-based isotherm, wt.%')
ax.legend(bbox_to_anchor=(1.1, 1.05))
plt.show()
```
