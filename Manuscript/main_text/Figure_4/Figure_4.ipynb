```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../Figure_3/result_analysis_time.csv')

print(df)
```

### measurement time: equilibrium vs kinetics

```python
text_size = 18
fig, ax = plt.subplots(1, 1, figsize=(15, 7.5))
ax.bar(df.name, df.isotherm_time, label='equilibrium')
ax.bar(df.name, df.kinetics_time, label='kinetics')
ax.set_xlabel('material', fontsize=text_size)
ax.set_ylabel('measurement time, min', fontsize=text_size)
ax.tick_params(axis='x', rotation=90, labelsize=text_size)
ax.tick_params(axis='y', labelsize=text_size)
#ax.legend()
fig.tight_layout()
plt.savefig('Figure_4.png', dpi=250)
plt.show()
```
