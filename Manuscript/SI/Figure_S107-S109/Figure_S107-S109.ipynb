```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
df_input = pd.read_csv('input_data_PSD_plot.csv')
print(df_input.head())
```

```python
for name in df_input.name.to_list():
    fig = plt.figure(figsize=(6, 2))
    ax = plt.subplot(111)
    current_names = df_input[df_input.name == name]
    
    df1 = pd.read_csv(current_names.path_kinetics.values[0]+current_names.filename_kinetics.values[0], 
                      skiprows=1, names=['A', 'B', 'C'])
    print(df1.head())
    df1 = df1[df1.B>=0]
    lns1 = ax.plot(df1.iloc[:, 0], df1.iloc[:, 1],label=name + '\nSKID', c='red')
    
    df2 = pd.read_csv(current_names.path_isotherm.values[0]+current_names.filename_isotherm.values[0], 
                      skiprows=1, names=['A', 'B', 'C'])
    print(df2.head())
    df2 = df2[df2.B>=0]
    lns2 = ax.plot(df2.iloc[:, 0], df2.iloc[:, 1],label=name + '\nequilibrium-based', c='blue')
    ax2 = ax.twinx()
    
    df3 = pd.read_csv(current_names.path_zeo.values[0]+current_names.filename_zeo.values[0], skiprows=11, 
                      sep='\s', names = ['Bin', 'Count', 'Cumulative_dist', 'Derivative_dist'])
    print(df3.head())
    lns3 = ax2.plot(df3.Bin/10, df3.Count,label='From SC: ' + current_names.filename_zeo.values[0].split('.')[0], c='green')
    ax.set_xlim([0, 2])
    ax.set_xlabel('Pore width, nm')
    ax.set_ylabel('Distribution, dV/dw')
    ax2.set_ylabel('Count', c='green')
    #ax.legend(bbox_to_anchor=(1.8, 1.00))
    #ax2.legend(bbox_to_anchor=(1.8, 0.1))
    # added these three lines
    lns = lns1+lns2+lns3
    labs = [l.get_label() for l in lns]
    ax2.legend(lns, labs, bbox_to_anchor=(1, 1.00))
    fig.tight_layout()
    plt.savefig('PSD_'+name+'.png', dpi=300)
    plt.show()
```
